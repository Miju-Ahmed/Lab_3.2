```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# convolution
signal = np.array([1, 2, 3, 4, 5])
impulse_signal = np.array([2, 3, 4])

convo_len = len(signal) + len(impulse_signal) - 1
convo_renge = []

for i in range(0, convo_len):
    convo_renge.append(i)


convolution = [0] * convo_len

for i in range(len(signal)):
    for j in range(len(impulse_signal)):
        convolution[i + j] += signal[i] * impulse_signal[j]
print(convolution)

plt.subplot(2, 2, 1)
plt.xlabel("range")
plt.ylabel("signal")
plt.stem(np.array([1, 2, 3, 4, 5]), signal)


plt.subplot(2, 2, 2)
plt.xlabel("range")
plt.ylabel("impulse signal")
plt.stem(np.array([1, 2, 3]), impulse_signal)

plt.subplot(2, 2, 3)
plt.xlabel("convo_range")
plt.ylabel("convolution signal")
plt.stem(np.array(convo_renge), convolution)

plt.show()
```

```python
# correlation : same as convolution except impulse_signal (flip())

impulse_signal = np.flip(impulse_signal)
print(impulse_signal)

convolution1 = [0] * convo_len

for i in range(len(signal)):
    for j in range(len(impulse_signal)):
        convolution1[i + j] += signal[i] * impulse_signal[j]
print(convolution1)

plt.xlabel("convo_range")
plt.ylabel("convolution signal")
plt.stem(np.array(convo_renge), convolution1)
plt.show()
```
