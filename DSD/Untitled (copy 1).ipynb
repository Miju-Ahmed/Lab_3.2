```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
f1 = 40
f2 = 80
f3 = 160
a1 = 10
a2 = 20
a3 = 40
fs = 512
```

```python
t = np.linspace(0,.05,int(fs),endpoint = True)

signal = a1*np.sin(2*np.pi*f1*t) + a2*np.sin(2*np.pi*f2*t) + a3*np.sin(2*np.pi*f3*t)

plt.figure(figsize = [20,4])
plt.plot(t,signal)
plt.title("Composite Signal")
plt.xlabel("Time")
plt.ylabel("Result")
plt.grid(True)
plt.show(True)
```

```python
N = 512
x = np.zeros(N)
for n in range(N):
    x[n] = a1 * np.sin(2*np.pi*f1 * n/fs) + a2 * np.sin(2*np.pi*f2 * n/fs) + a3 * np.sin(2*np.pi*f3 * n/fs)
    
X = np.zeros(N, dtype=complex)
for k in range(N):
    for n in range(N):
        X[k] = x[n]*np.exp(-2j*np.pi*n*k/N)
    X[k] = round(X[k].real,4) + 1j*round(X[k].imag,4)
#print(X)
        
```

```python
#Calculate Magnitude

xm = []
N=512
for k in range(N):
    a = np.sqrt(X[k].real**2 + X[k].imag**2)
    xm.append(a)

    
plt.figure(figsize = [10,4])
plt.plot(np.arange(512),xm)
plt.show()

#print(xm)
```

```python
#Calculate Phase Spectrum

xp = []
N=512
for k in range(N):
    a = math.atan2(X[k].imag,X[k].real)
    a = np.degrees(a)
    if a==180:
        a=0
    xp.append(a)
plt.figure(figsize = [10,4])
plt.plot(np.arange(512),xp)
plt.grid(True)
plt.show()
```
