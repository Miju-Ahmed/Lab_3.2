```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
f1 = 1000
f2 = 2000
ph2 = (3 * np.pi) / 4
amp2 = 0.5
sampling_rate = 8000
time = np.linspace(0, 0.001, sampling_rate)
signal = np.sin(2 * np.pi * 1000 * time) + amp2 * np.sin((2 * np.pi * 2000 * time) + ph2)

plt.plot(time, signal)
plt.grid(True)
plt.show()
print(ph2)
```

```python
x = []
ts = 1 / sampling_rate
for i in range(0, 8):
    sg = np.sin(2 * np.pi * f1 * i * ts) + amp2 * np.sin((2 * np.pi * f2 * i * ts) + ph2)
    sg = float("{:.3f}".format(sg))
    x.append(sg)

plt.stem(np.arange(8), x)
print(0.5 * np.sin(ph2))
```

```python
Xr = []
Xim = []
for j in range(0, 8):
    a = 0
    b = 0
    for i in range(0, 8):
        a = a + float("{:.3f}".format((x[i] * np.cos( 2 * np.pi *(j)* i / 8))))
        b = b - float("{:.3f}".format((x[i] * np.sin( 2 * np.pi *(j)* i / 8))))
    Xr.append(float("{:.3f}".format(a)))
    Xim.append(float("{:.3f}".format(b)))


    
print(Xr)
print(Xim)
    
```

```python
# Magnitude spectrum
Xm = []
for i in range(0, 8):
    a = np.sqrt((Xr[i] ** 2) + Xim[i] **2);
    Xm.append(a)
plt.stem(np.arange(0, 8), Xm)
```

```python
# Phase spectrum
phase = []
for i in range(0, 8):
    angle = math.atan2(Xim[i], Xr[i])
    angle = math.degrees(angle)
    if angle == 180: #tan 180 = tan 0 
        angle = 0
    phase.append(angle)
plt.stem(np.arange(0, 8), phase)
    
```

```python
# spectrrm of real value of X(m)
plt.stem(np.arange(0, 8), Xr)
```

```python
# spectrum of imaginary value of X(m)
plt.stem(np.arange(0, 8), Xim)
```

```python
# Spectrum inverse X(M)/x(n)
Ire = []
Iim = []

for i in range(0, 8) :
    a = 0
    b = 0
    for j in range(0, 8):
        a = a + float("{:.3f}".format(Xr[j] * np.cos(2 * np.pi * i * j / 8))) - float("{:.3f}".format(Xim[j] * np.sin(2 * np.pi * i * j / 8)))
        b = b + float("{:.3f}".format(Xr[j] * np.sin(2 * np.pi * i * j / 8))) + float("{:.3f}".format(Xim[j] * np.cos(2 * np.pi * i * j / 8)))
    a = a / 8
    b = b / 8
    Ire.append(a)
    Iim.append(b)
    

plt.stem(np.arange(8), Ire)
print(Ire)
print(x)
print(Iim)
```
