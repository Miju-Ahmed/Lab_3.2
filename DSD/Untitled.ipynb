```python
import numpy as np
import matplotlib.pyplot as plt

f1= 1000
f2 = 2000
amplitude = 0.5
phase = (3*np.pi/4)
fs = 8000


t = np.linspace(0,0.001, int(fs), endpoint=False)

sine_wave = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2*t + phase)

#plt.subplot(2,2,1)
plt.plot(t,sine_wave,color='red')
plt.title("Sine Wave")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
x = np.zeros(8)
fs = 8000
ts = 1/fs
for n in range(0,8):
    x[n] = np.sin(2*np.pi*f1*n*ts) + 0.5 * np.sin(2*np.pi*f2*n*ts + 3*np.pi/4)

sig_range = np.arange(8)    
#plt.subplot(2,2,2)
plt.stem(sig_range, x)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)

print(x)
```

```python
X = np.zeros(8, dtype=complex)
for k in range(8):
    for n in range(8):
        X[k] += x[n]*np.cos(n*np.pi*k/4) - x[n]*1j*np.sin(n*np.pi*k/4)
    X[k] = round(X[k].real, 2) + 1j*round(X[k].imag, 2)
frequencies = np.fft.fftfreq(8, d=ts)


#plt.subplot(2, 1, 2)
plt.stem(frequencies, np.abs(X))
plt.title("Magnitude spectrum of DFT")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.tight_layout()
plt.show()
print(X)
```

```python
#Magnitude Spectrum

Xm = []
for i in range(8):
    a = np.sqrt(X[i].real**2 + X[i].imag**2)
    Xm.append(a)
print(Xm)
plt.stem(np.arange(0,8), Xm)
plt.title("Magnitude Spectrum of DFT")
plt.show()
```

```python
Xpo = []
for i in range(8):
    a = Xm[i]**2
    Xpo.append(a)
plt.stem(np.arange(0,8), Xpo)
plt.title("Power Spectrum of DFT")
plt.tight_layout()
plt.show()
```

```python
#Phase Spectrum

Xp = []
for i in range(8):
    a = np.arctan2(X[i].imag,X[i].real)
    Xp.append(a)

plt.stem(np.arange(0,8), Xp)
plt.title("Phase Spectrum of DFT")
plt.tight_layout()
plt.show()
```
